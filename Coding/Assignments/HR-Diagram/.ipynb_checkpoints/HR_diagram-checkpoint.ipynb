In this assignment, we will be importing data from a `.txt` file and plotting the data.

The data we are observing today is `StarData`, which contains;
* *Column 1*: The Star ID number
* *Column 2*: The apparent $V$ Magnitude, $m_V$
* *Column 3*: The observed $B-V$ colour, or $m_B - m_V$
* *Column 4*: The observed *Parallax* in arcseconds
* *Column 5*: The uncertainty in the obeserved *Parallax*

My code will import the data from the `.txt` file and sort it into specific arrays, and then plot said data in hopes of recreating the ***Hertzsprung-Russell Diagram***.

To do this, we will need to compute the **Absolute Magnitude $M_V$** and compare this to the observed **$B-V$** colour.

To calculate **$M_V$**, we will use the formula
### $$ m_V - M_V = 5 log(d/10) $$
or
### $$ M_V = m_v - 5 log(d/10) $$
Where
### $$ d=1/p $$

Here:
- $p$ is the *Parallax angle*
- $d$ is the distance in *parsecs*

Using all this information, I will plot a diagram of $M_V$ vs $B-V$ Colour Index, and this should result in a diagram similar to the Hertzsprung Russell Diagram

```python
## Import neccesary modules for assignment
import matplotlib.pyplot as plt
import numpy as np
```

## Task 1:
To begin, I will start by checking that the `.txt` file is loaded correctly.

```python
## I use the simplest method of opening and obtaining the text from the file
FileData = open('StarData.txt', 'r')

## I print the first line of the .txt file
print(FileData.readline())

## Assign the entire collection of lines to some variable
StarData = FileData.readlines()

## Begin loop that assigns a variable to a line
for line in StarData:
    ## Split lines using Python's command. Here it splits where it seems best to split, usually at spaces
    columns = line.split()

## Close File
FileData.close()

## Return earlier variables for user to help understanding
print(line)
print(columns)
```

## Task 2:
Using a more compact method, I will extract the data from the `.txt` file and assign the data to its respective arrays.

Here we will be observing a plot of the $m_v$ vs $B-V$ colour index

```python
## I now try a more advanced method.
##  A benefit of the 'with' statement is that it will close the file after it is complete
with open('StarData.txt') as StarData:
    
    ## Similar to before, except this time I split each line up and place it into a list
    lines = StarData.read().splitlines()

    ## Number of spaces needed for my arrays. The '// 2' here comes from
    ##  the extra space at the end of each line, creating a 'phantom line' of sorts
    n = (len(lines) + 1) // 2

    ## Initiate the arrays for each column listed above in the introduction
    starID, app_V_magn, obs_B_V = np.zeros(n), np.zeros(n), np.zeros(n)
    obs_para, obs_para_uncertainty = np.zeros(n), np.zeros(n)
    
    ## Begin for loop to assign each value in the column to the respective column
    ##  I use `enumerate()` to obtain the location AND the data in each 'line'
    for i, line in enumerate(lines):

        ## In reference to earlier, this removes the 'phantom line'
        lines.pop(0)

        ## Splits list into multiple components
        columns = line.split()

        ## Assigns respective values to each array
        starID[i] = float(columns[0])
        app_V_magn[i] = float(columns[1])
        obs_B_V[i] = float(columns[2])
        obs_para[i] = float(columns[3])
        obs_para_uncertainty[i] = float(columns[4])
    
## Begins plot definitions for Apparent Magnitude vs Observed B-V
xmin, xmax = -0.5, 2.5

plt.figure(1, figsize=(10,10)).set_facecolor('black')
plt.rcParams['axes.facecolor'] = 'black'
plt.scatter(obs_B_V, app_V_magn, s=3, c=obs_B_V, cmap='Spectral_r', vmin=np.min(obs_B_V), vmax=2)
plt.title("Apparent Magnitude ($m_V$) vs B-V Colour ($m_B - m_V$)", c='w')
plt.xlabel("$m_B - m_V$", c='w')
plt.ylabel("Apparent Magnitude ($m_V$)", c='w')
plt.xlim(xmin, xmax)
plt.tick_params(colors='w')

plt.show(1)
```

## Task 3:
A more useful and practical version of the above diagram is one that compares the **Absolute Magnitude $M_V$** to the **B-V** colour index. To do this, I will be using the formula stated in the introduciton.

Once I have obtained **$M_V$** I will plot it against $B-V$.

```python
## Using the arrays from above, I can convert the parallax into a parsec array
d = 1 / obs_para

## Using the formula in the introduction, I calculate the Absolute Magnitude
abs_V_magn = app_V_magn - 5 * (np.log10(d)) + 5

## Begin plot definitions
xmin, xmax = -0.5, 2.5
ymin, ymax = 20, -10

plt.figure(2, figsize=(15,15)).set_facecolor('black')
plt.rcParams['axes.facecolor'] = 'black'
plt.scatter(obs_B_V, abs_V_magn, s=3, c=obs_B_V, cmap='Spectral_r', vmin=np.min(obs_B_V), vmax=2)
plt.title("Absolute Magnitude ($M_V$) vs B-V Colour ($m_B - m_V$)", c='w')
plt.xlabel("$m_B - m_V$", c='w')
plt.ylabel("Absolute Magnitude ($M_V$)", c='w')
plt.ylim(ymin, ymax)
plt.xlim(xmin, xmax)
plt.tick_params(colors='w')
plt.show(2)
```


Using this graph, we can see certain characteristics of the diagram. There appears to be a winding line going diagonally down through the centre of the plot, a large 'clump' that branches off at around midway down this winding line, and a skinny branch far seperated from the central line in the bottom left corner.

We can classify these sections into the '**Main Sequence Stars**', '**Red Giant Stars**', and '**White Dwarf Stars**' respectively.

To try and count the amount of stars in these sections, I use a crude method of boxing off the data. This should obtain a rough estimate for each category.

```python
## Begin by initialising counters for each category
rGiants = 0 ##Red Giants
wDwarfs = 0 ##White Dwarfs
stars = len(obs_B_V) ## Total amount of stars

## Limits for where Red Giants begin and end, and similarly for White Dwarfs begin and end
B_V_lim = [0.8, 1.2]
magn_lim = [3, 9]

## Begin `for` loop to look at each star in the system
for i in range(stars):
    
    ## Checks if the stars match each condition. If True, then a count is added
    if obs_B_V[i] > B_V_lim[0] and abs_V_magn[i] < magn_lim[0]:
        rGiants += 1
    if obs_B_V[i] <  B_V_lim[1] and abs_V_magn[i] > magn_lim[1]:
        wDwarfs += 1
        
## Calculates Main Sequence Stars
mainSequence = stars - rGiants - wDwarfs

## Relays information back to user
print("""Red Giants in data:               {0:4.0f} ({1:3.1f}%)
White Dwarfs in data:             {2:4.0f} ({3:3.1f}%)
Main Sequence stars in data:      {4:4.0f} ({5:4.1f}%)""".format(rGiants, rGiants/stars * 100, wDwarfs, wDwarfs/stars * 100, mainSequence, mainSequence/stars * 100))


## Begin plot definitions
xmin, xmax = -0.5, 2.5
ymin, ymax = 20, -10

plt.figure(3, figsize=(15,15)).set_facecolor('black')
plt.rcParams['axes.facecolor'] = 'black'
plt.scatter(obs_B_V, abs_V_magn, s=3, c=obs_B_V, cmap='Spectral_r', vmin=np.min(obs_B_V), vmax=2)
plt.title("Absolute Magnitude ($M_V$) vs B-V Colour ($m_B - m_V$)", c='w')
plt.xlabel("B-V Colour ($m_B - m_V$)", c='w')
plt.ylabel("Absolute Magnitude ($M_V$)", c='w')
plt.ylim(20, -10)
plt.xlim(-0.5, 2.5)
plt.tick_params(colors='w')

## Shows user the 'boxes' that were used to count the amount of stars
plt.axhline(magn_lim[0], xmin=0.434, linewidth=2, c='w')
plt.axvline(B_V_lim[0], ymin=0.568, linewidth=2, c='w')
plt.annotate(s='Red Giants\n $\sim$ {}'.format(rGiants), xy=(1.5, -6), fontsize=30, c='w')

plt.axhline(magn_lim[1], xmax=0.566, linewidth=2, c='w')
plt.axvline(B_V_lim[1], ymax=0.366, linewidth=2, c='w')
plt.annotate(s='White Dwarfs\n $\sim$ {}'.format(wDwarfs), xy=(-0.4,17), fontsize=30, c='w')

plt.annotate(s='Main Sequence\n $\sim$ {}'.format(mainSequence), xy=(1.3, 16), fontsize=30, c='w')

plt.show(3)
```

## Task 4:
From prior knowledge, we know that Absolute Magnitude is directly proportional to **Luminosity**.  

Using the formula:

$$ M_V = -2.5 log_{10}\frac{L_V}{L_0} $$
or
$$ \frac{L_V}{L_0} \simeq 10^{-0.4 M_V} $$

Where
- $M_V$ is the Absolute Magnitude of the star we are observing
- $L_V$ is the *Luminosity* of the star we are observing
- $L_0$ is the *Luminosity* of the star we are comparing to, which in this case is the Sun.

Since we know the *luminosity* of the Sun, we are instead going to compare the Luminosity of the star to the *luminosity* of the Sun.

The following cell was used for figuring out the Luminosity part of Task 4, and getting a second axis to work,

I have left it here for educational purposes.

```python
## Compute the Luminosity of each Star
Lum = 10 ** (-0.4 * abs_V_magn)

## Begin subplot definitions needed for secondary axis
fig, ax1 = plt.subplots(figsize=(5,5))

fig.set_facecolor('black')
ax1.set_xlabel("B-V Colour ($m_B - m_V$)", c='w')
ax1.set_ylabel("Absolute Magnitude ($M_V$)", c='w')
ax1.scatter(obs_B_V, abs_V_magn, s=3, c='b')#obs_B_V, cmap='Spectral_r', vmin=np.min(obs_B_V), vmax=2)
ax1.tick_params(labelcolor='w')
ax1.set_ylim(20,-10)
ax1.set_xlim(-0.5, 2.5)


## Begin defining a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

ax2.set_ylabel("Luminosity", c='w')
ax2.scatter(obs_B_V, Lum, s=3, c='r')#obs_B_V, cmap='hot_r', vmin=np.min(obs_B_V), vmax=2)
ax2.tick_params(axis='y', labelcolor='w')#
ax2.set_yscale('log')

## I had to manually change the limits to get this to fit right
##  To see this best, 10^(-8) to 10^(4) is the best fit
##  Red is the Luminosity, Blue is the Abs Magn
ax2.set_ylim(10**(-6), 10**(4))

plt.show()
```
